```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#import sklearn
```

```python
collisions = pd.read_csv("Motor_Vehicle_Collisions_-_Crashes.csv")
```

```python
#Converting Crash Date To DateTime, and Splitting into Day Month Year
collisions['CRASH DATE']= collisions['CRASH DATE'].astype('datetime64[ns]')


collisions['day']= collisions['CRASH DATE'].dt.day
colday = collisions['day']
collisions['month']= collisions['CRASH DATE'].dt.month
colmonth = collisions['month']
collisions['year']= collisions['CRASH DATE'].dt.year
colyear = collisions['year']


collisions.insert(0, column="YEAR", value=colyear)
collisions.insert(1, column="MONTH", value=colmonth)
collisions.insert(2, column="DAY", value=colday)

collisions.info
```

```python
collisions['CRASH TIME']= collisions['CRASH TIME'].astype('datetime64[ns]')
collisions.head


colhr = collisions['hour'] = collisions['CRASH TIME'].dt.hour
colmin = collisions['minute'] = collisions['CRASH TIME'].dt.minute

collisions.insert(4, column="HOUR", value=colhr)
collisions.insert(5, column="MINUTE", value=colmin)
```

```python
sns.barplot(x = collisions.groupby("BOROUGH").size().index,
            y = collisions.groupby("BOROUGH").size().values)
plt.title("No. OF ACCIDENTS PER BOROUGH")
```

```python
col_bor1 = collisions.groupby("BOROUGH")[["NUMBER OF PERSONS INJURED", "NUMBER OF PERSONS KILLED"]].sum()

fig, ax = plt.subplots(1, 2, figsize = (15,5))

ax[1].set_xticklabels(labels = col_bor1.index)
ax[0].set_xticklabels(labels = col_bor1.index)

sns.barplot(col_bor1.index, col_bor1["NUMBER OF PERSONS INJURED"], ax = ax[0])
sns.barplot(col_bor1.index, col_bor1["NUMBER OF PERSONS KILLED"], ax = ax[1])

plt.suptitle("Injured vs Killed in NYC")
```

```python
collisions_injured = collisions[[i for i in collisions.columns for c in ["NUMBER OF PEDESTRIANS INJURED", "NUMBER OF CYCLIST INJURED", "NUMBER OF MOTORIST INJURED"] if c in i]].sum()
collisions_killed = collisions[[i for i in collisions.columns for c in ["NUMBER OF PEDESTRIANS KILLED", "NUMBER OF CYCLIST KILLED", "NUMBER OF MOTORIST KILLED"] if c in i]].sum()
collisions_injured.index = ["Pedestrian", "Cyclist", "Motorist"]
collisions_killed.index = ["Pedestrian", "Cyclist", "Motorist"]

fig, ax = plt.subplots(1,2,figsize=(14,5))
plt.suptitle("KILLED vs INJURED PER ACCIDENT", fontsize = 20, x = 0.5, y = 1.02)
ax[0].set_title("INJURED", fontsize = 14)
ax[1].set_title("KILLED", fontsize = 14)

sns.barplot(collisions_injured.index, collisions_injured.values, ax = ax[0])
sns.barplot(collisions_killed.index, collisions_killed.values, ax = ax[1])
```
